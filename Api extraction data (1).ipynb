```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd

# Set up client credentials flow
client_id = 'client id'
client_secret = 'clientsecret'
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Replace this with the Spotify URI of the resource you want recommendations for
playlist_uri = 'spotify:playlist:37i9dQZEVXbNG2KDcFcKOF'

# Extract track information from the playlist
results = sp.playlist_tracks(playlist_uri)

# Initialize lists to store track, artist, and album information
tracks_data = []
artists_data = []
albums_data = []

# Iterate over the tracks and extract relevant information
for track in results['items']:
    track_info = track['track']
    
    # Track data
    track_id = track_info['id']
    track_name = track_info['name']
    track_popularity = track_info['popularity']
    tracks_data.append({'track_id': track_id, 'track_name': track_name, 'track_popularity': track_popularity})
    
    # Artist data
    artists_info = track_info['artists']
    for artist_info in artists_info:
        artist_id = artist_info['id']
        artist_name = artist_info['name']
        artist_popularity = sp.artist(artist_id)['popularity']
        followers = sp.artist(artist_id)['followers']['total']
        genre = sp.artist(artist_id)['genres']
        artists_data.append({'artist_id': artist_id, 'artist_name': artist_name, 
                             'artist_popularity': artist_popularity, 'followers': followers, 'genre': genre})
    
    # Album data
    album_info = track_info['album']
    album_id = album_info['id']
    album_name = album_info['name']
    release_date = album_info['release_date']
    album_popularity = sp.album(album_id)['popularity']
    for artist_info in artists_info:
        artist_id = artist_info['id']
        albums_data.append({'album_id': album_id, 'album_name': album_name, 'release_date': release_date,
                             'album_popularity': album_popularity, 'artist_id': artist_id})

# Create DataFrames from the lists
tracks_df = pd.DataFrame(tracks_data)
artists_df = pd.DataFrame(artists_data)
albums_df = pd.DataFrame(albums_data)

# Save DataFrames as CSV files
tracks_df.to_csv('tracks.csv', index=False)
artists_df.to_csv('artists.csv', index=False)
albums_df.to_csv('albums.csv', index=False)

print("CSV files saved successfully.")
```
